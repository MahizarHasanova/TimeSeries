```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

import io

# Read the file in binary mode
with open('AdventureWorks2014.csv', 'rb') as f:
    content = f.read()

# Try decoding with different encodings
encodings_to_try = ['utf-8', 'latin-1', 'utf-16', 'utf-16-le', 'utf-16-be']
decoded_content = None

for encoding in encodings_to_try:
    try:
        decoded_content = content.decode(encoding)
        break
    except UnicodeDecodeError:
        continue

if decoded_content is None:
    print("Unable to decode the file using any of the tried encodings.")
else:
    # Remove BOM if present
    decoded_content = decoded_content.lstrip('\ufeff')
    # Create a DataFrame from the decoded content
    df = pd.read_csv(io.StringIO(decoded_content))

# Now you can work with the DataFrame 'df'
```

```python
def linreg(X, Y):
    """
    return a,b in solution to y = ax + b such that root mean square distance between trend line and original points is minimized
    """
    N = len(X)
    Sx = Sy = Sxx = Syy = Sxy = 0.0
    for x, y in zip(X, Y):
        Sx = Sx + x
        Sy = Sy + y
        Sxx = Sxx + x*x
        Syy = Syy + y*y
        Sxy = Sxy + x*y
    det = Sxx * N - Sx * Sx
    return (Sxy * N - Sy * Sx)/det, (Sxx * Sy - Sx * Sxy)/det
```
